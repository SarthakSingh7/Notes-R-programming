## Coerce or convert the data type of an object to another
This task is interesting where you can change or convert the data type of one object to another. To perform this action you have to use the prefix as. before the data type as the command. The syntax for doing that is as.data_type() of the object which you want to coerce.

```r

# A simple R program convert data type of an object to another
 
# Logical
print(as.numeric(TRUE))
 
# Integer
print(as.complex(3L))
 
# Numeric
print(as.logical(10.5))
 
# Complex
print(as.character(1+2i))
 
# Can't possible
print(as.numeric("12-04-2020"))

#Note: All the coercions are not possible and if attempted will be returning an NA value.
```

## Explicit Coercion in R Programming

Coercing of an object from one type of class to another is known as explicit coercion. It is achieved through some functions which are similar to the base functions. But they differ from base functions as they are not generic and hence do not call S3 class methods for conversion.

Difference between conversion, coercion and cast:

Normally, whatever is converted implicitly is referred to as coercion and if converted explicitly then it is known as casting. Conversion signifies both types- coercion and casting.

Explicit coercion to character
There are two functions to do so as.character() and as.string(). 

If v is a vector which is needed to be converted into character then it can be converted as:

 as.character(v, encoding = NULL)
 as.string(v, encoding = NULL)
Here encoding parameter informs R compiler about encoding of the vector and helps internally in managing character and string vectors.

# Reshaping o Recasting dataframe






## variable or feature & value

### What does Recasting of a dataframe mean?
Recasting is the process of manipulating a dataframe in terms of its variables 

### What is the need of Recasting the dataframe? 

1. Recasting helps in reshaping the data

2. recasting of reshaping could bring more insights from the data when it is seen from a different perspective


```r
# data frame example 2
pd = data.frame('Name' = c('Senthil', 'Senthil', 'Sam', 'Sam'), 
               'Month' = c('Jan', 'Feb', 'Jan', 'Feb'),
               'BS' = c(141.2,139.3,135.2,160.1),
               'BP' = c(90,78,80,81))
print(pd)
```

### Recast can be done in two steps
Step 1: Melt

Step 2: Cast

We can recast the dataframe into another form 

When we want to use **melt** and **cast** commands to recast the dataframe for that we need to identify what are called ***identifier variables*** and ***measurement variables*** of your data frame. 

The **rules** for identifying this:

1. **identifier variables** are: most of the discrete type variables can be identifier variables 
2. measure the numeric variables can be **measurement variables** and there are certain rules for the **measurement variables** such as: categorical and date variables **cannot** be measurement variables. 

So, the keyidea is: from the data frame, you have to identify what are called identifier variables and what are called measurement variables Once you have identify this identifier variables and measurement variables, you are ready to do the ***melt operation*** 


### Using library() command for reshape2 library

1. The melt command is available in the reshape2 library

2. Call the library 'reshape2' using the library() command

3. for the melt command we have to give the dataframe as the first argument, we have to specify what are the identifier variables in the dataframe and specify what are the mesurement variables in the dataframe.



```r
# step 1: Melt

library(reshape2)   # importing the library or opening a library reshape2

df = melt(pd, id.vars = c('Name', 'Month'), measure.vars = c('BS','BP'))

print(df)

# NOTES:  
# here Name and Month are identifier variables
# and BS and BP are measurement variables 

# Name, Month --- Identifier variables 

# BS, BP --- Measure variables
```

When the melt command is executed, it will take this identifier variable and keep them as it is and 
it will convert the measured variables into one single column named (variable) and stores the values of those
variables in another colum named (value)


in simple words 

The column that carries the **measurement variables** is named as **variable** 

and the columns that holds the values of the measurement variables is named as **Values**

other columns are for **identifier variables**

Step 1:  **Melt**
identify the **'identifying variables'** and **'measurement variables'** of the dataframe and use the melt command 
to melt the data frame to get this structure

step 1 completed 



next step

Step 2: **Cast**
since we are using dataframe we will use **dcast()** function this is available in the **reshape2** library

the syntax of decast is as follows:
1. the dcast command takes the **dataframe as first attribute** the dataframe which we want to decast
2. the second attribute is the **formula** which will be implemented
3. the 3rd attribute is **value.var = col.with values** we have to specify the columns from which the values to be taken from when we are dcasting

### **<div align = 'center'>dcast(data ,formula, value.var = col.with values) </div>**



```r
# Step 2: Cast

# we are taking the data frame which we melt df

df2 = dcast(df, variable+Month ~ Name, value.var = 'value')
print(df2)

# formula explained
#                                        variable + Month ~ Name
# We want to have variable and month columns as constant and the categoies in the Name become new Variable columns


## third attribute 
#                                        value.var = 'value'
# the values for those 2 names columns should be picked from the column 'value'
```

1. the recast function take the **data** or **dataframe** as the first attribute
